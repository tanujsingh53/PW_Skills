# Statistics Advance Part 1

 1. What is a random variable in probability theory?

A random variable is a variable that assigns a numerical value to the outcome of a random experiment.

 2. What are the types of random variables?
- Discrete Random Variable
- Continuous Random Variable

3. What is the difference between discrete and continuous distributions?
| Type      | Values | Example                 |
|-----------|--------|--------------------------|
| Discrete  | Countable | Number of heads in 5 tosses |
| Continuous| Any value in range | Heights of students       |


 4. What are probability distribution functions (PDF)?
- **PDF** defines the probability density for continuous random variables.
- For discrete variables, it's called **PMF** (Probability Mass Function).


5. How do cumulative distribution functions (CDF) differ from PDF?
- **PDF**: Gives density
- **CDF**: Gives cumulative probability up to a point: `P(X ≤ x)`


6. What is a discrete uniform distribution?

All outcomes have equal probability. Example: rolling a fair die.

7. What are the key properties of a Bernoulli distribution?
- Two outcomes: 1 (success), 0 (failure)
- Mean = p, Variance = p(1-p)


8. What is the binomial distribution, and how is it used?
- Repeated Bernoulli trials (n)
- Formula: C(n, k) * p^k * (1-p)^(n-k)
- Use: Number of successes in fixed trials


9. What is the Poisson distribution and where is it applied?
- Used for rare events in time/space
- Formula: e^(-λ) * λ^k / k!


10. What is a continuous uniform distribution?
- Equal probability for all values in range [a, b]
- PDF = 1 / (b - a)

11. What are the characteristics of a normal distribution?
- Bell-shaped, symmetric
- Mean = Median = Mode


12. What is the standard normal distribution, and why is it important?
- Mean = 0, Std Dev = 1
- Basis for Z-score and many tests


13. What is the Central Limit Theorem (CLT)?
- The sampling distribution of the mean becomes normal as n → ∞.

14. How does CLT relate to normal distribution?

Even if population is not normal, sample mean tends to normal with large n.

15. What is the application of Z statistics in hypothesis testing?
- Used when population standard deviation is known
- Formula: (x̄ - μ) / (σ / √n)


16. How do you calculate a Z-score?

Formula: Z = (X - μ) / σ


17. What are point and interval estimates?
- Point estimate: single value
- Interval estimate: range with confidence level


18. What is the significance of confidence intervals?

CI tells the range in which the population parameter is likely to lie.


19. What is the relationship between a Z-score and a CI?

CI uses Z-score for margin of error:
x̄ ± Z * (σ / √n)

20. How are Z-scores used to compare distributions?

They standardize values across different distributions.


21. What are the assumptions for CLT?
- Random, independent samples
- Finite mean and variance
- Sample size ≥ 30


 22. What is the expected value in a probability distribution?

Weighted average of all outcomes:
E(X) = Σ P(x) * x

23. How does a probability distribution relate to expected value?

It defines the likelihood of outcomes used to compute expected value.

Q3. Write a Python function to calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.


```python
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    return {0: bernoulli.pmf(0, p), 1: bernoulli.pmf(1, p)}

print("Bernoulli PDF (p=0.6):", bernoulli_pdf(0.6))
```

Q4. Simulate a binomial distribution with n = 10, p = 0.5 and plot its histogram.

```python
from scipy.stats import binom
import numpy as np

data = binom.rvs(n=10, p=0.5, size=1000)

plt.hist(data, bins=range(12), edgecolor='black', align='left')
plt.title("Binomial Distribution Histogram")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

Q5. Create a Poisson distribution and visualize it using Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define lambda
lam = 4

# Generate x values and compute PMF
x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=lam)

# Plot (fixed without use_line_collection)
markerline, stemlines, baseline = plt.stem(x, pmf)
plt.setp(markerline, 'markerfacecolor', 'blue')
plt.title("Poisson Distribution (λ=4)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

Q6. Write a Python program to calculate and plot the Cumulative Distribution Function (CDF) of a discrete uniform distribution

```python
from scipy.stats import randint

x = np.arange(1, 7)
cdf = randint.cdf(x, 1, 7)

plt.step(x, cdf, where='mid')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

Q12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).

```python
x = np.linspace(-4, 4, 1000)
pdf = np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)

plt.plot(x, pdf)
plt.title("Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

Q14. Calculate the Z-score for a given data point and compare it to a standard normal distribution.

Q15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
sample = np.random.normal(100, 10, 30)
pop_mean = 105
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
z_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))

print("Z-statistic:", z_stat)
```

 Q16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import scipy.stats as stats

conf_int = stats.t.interval(0.95, len(sample)-1, loc=sample_mean, scale=sample_std/np.sqrt(len(sample)))
print("95% Confidence Interval:", conf_int)
```

Q17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
data = np.random.normal(50, 5, 100)
mean = np.mean(data)
std_err = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print("Mean:", mean)
print("95% Confidence Interval:", ci)
```
