# Statistics Part 2

1. What is hypothesis testing in statistics?

Answer: Hypothesis testing is a method used to make decisions about a population based on sample data. It checks whether a claim (hypothesis) about a population is true or not.


2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:

Null Hypothesis (H₀): A statement that there is no effect or no difference.

Alternative Hypothesis (H₁): A statement that there is an effect or a difference.
They are opposites. We test the null hypothesis, and if it's not supported by the data, we accept the alternative.

3. What is the significance level in hypothesis testing, and why is it important?

Answer: The significance level (usually written as α) is the chance of rejecting the null hypothesis when it is actually true.
Common values are 0.05 or 5%. It helps decide how strong the evidence needs to be to reject the null.

4. What does a P-value represent in hypothesis testing?

Answer: The p-value shows how likely the observed result is if the null hypothesis is true. A smaller p-value means stronger evidence against the null hypothesis.






5. How do you interpret the P-value in hypothesis testing?

Answer:
If p-value ≤ α, reject the null hypothesis (result is significant).
If p-value > α, do not reject the null hypothesis (result is not significant).





6. What are Type 1 and Type 2 errors in hypothesis testing?

Answer:

Type 1 Error (False Positive): Rejecting a true null hypothesis.

Type 2 Error (False Negative): Not rejecting a false null hypothesis.



7. What is the difference between a one-tailed and a two-tailed test?

Answer:

One-tailed test: Checks for effect in one direction (e.g., greater than).

Two-tailed test: Checks for effect in both directions (e.g., not equal to).

8. What is the Z-test, and when is it used?

Answer: A Z-test is used to compare a sample mean with a population mean when the population standard deviation is known and sample size is large (n > 30).

10. What is the T-distribution, and when should it be used instead of the normal distribution?

Answer: The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown.

11. What is the difference between a Z-test and a T-test?

Answer:

Z-test: Used when population standard deviation is known and large sample.

T-test: Used when population standard deviation is unknown and small sample.



12. What is the T-test, and how is it used in hypothesis testing?

Answer: A T-test is used to compare means (e.g., between groups) when the population standard deviation is unknown.

13. What is the relationship between Z-test and T-test?

Answer: Both compare means. The T-test is used for smaller samples, and the Z-test for larger samples. As sample size increases, the T-distribution becomes similar to the normal distribution.

14. What is a confidence interval, and how is it used?

Answer: A confidence interval gives a range of values within which we expect the true population value to fall, with a certain level of confidence (e.g., 95%).



15. What is the margin of error, and how does it affect the confidence interval?

Answer: The margin of error is the amount added or subtracted from the sample result to get the confidence interval. A larger margin of error means a wider interval.



16. How is Bayes' Theorem used in statistics, and what is its significance?

Answer: Bayes’ Theorem updates the probability of an event based on new evidence. It’s used in machine learning, medical testing, and decision making.



17. What is the Chi-square distribution, and when is it used?

Answer: The Chi-square distribution is used for testing relationships between categorical variables, like in independence or goodness-of-fit tests.

18. What is the Chi-square goodness of fit test, and how is it applied?

Answer: It checks if observed data fits a specific distribution. Example: Does the number of red, green, and blue balls match expected numbers?

19. What is the F-distribution, and when is it used?

Answer: The F-distribution is used to compare two variances. It's the basis for ANOVA tests.

20. What is an ANOVA test, and what are its assumptions?

Answer: ANOVA (Analysis of Variance) checks if the means of 3 or more groups are equal.
Assumptions: Normal distribution, equal variances, and independent samples.



21. What are the different types of ANOVA tests?

Answer:

One-way ANOVA: Tests one factor across groups.

Two-way ANOVA: Tests two factors and their interaction.



22. What is the F-test, and how does it relate to hypothesis testing?

Answer: The F-test compares variances to see if groups are significantly different. It is used in ANOVA and regression.

1. Q: Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
from scipy.stats import norm
import numpy as np

sample = [68, 70, 72, 71, 69]
sample_mean = np.mean(sample)
population_mean = 70
std_dev = 2  # known population std
n = len(sample)

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

print("Z-score:", z)
print("P-value:", p)
```

3. Q: Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
def one_sample_z_test(sample, pop_mean, pop_std):
    mean = np.mean(sample)
    n = len(sample)
    z = (mean - pop_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = np.random.normal(50, 10, 40)
z, p = one_sample_z_test(sample, 50, 10)
print("Z:", z, "| P-value:", p)
```

4. Q: Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import matplotlib.pyplot as plt

z = 2.0
alpha = 0.05
critical_z = norm.ppf(1 - alpha/2)

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -critical_z) | (x >= critical_z), color='red', alpha=0.5, label="Rejection Region")
plt.axvline(z, color='blue', linestyle='--', label=f"Z = {z}")
plt.title("Two-tailed Z-Test")
plt.legend()
plt.show()
```

5. Q: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
def plot_type_errors(mu0=0, mu1=1, sigma=1, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)

    z_crit = norm.ppf(1 - alpha)

    plt.plot(x, y0, label="H₀ distribution")
    plt.plot(x, y1, label="H₁ distribution")
    plt.fill_between(x, y0, where=(x > z_crit), color='red', alpha=0.5, label="Type I Error")
    plt.fill_between(x, y1, where=(x <= z_crit), color='blue', alpha=0.3, label="Type II Error")
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

plot_type_errors()
```

9. Q: Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    margin = se * stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    return mean - margin, mean + margin

data = np.random.normal(50, 10, 50)
ci = confidence_interval(data)
print("Confidence Interval:", ci)
```

11. Q: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prior_A * prob_B_given_A) / prob_B

# Example: Disease test
prior_disease = 0.01
prob_pos_given_disease = 0.99
prob_pos = 0.05

posterior = bayes_theorem(prior_disease, prob_pos_given_disease, prob_pos)
print("Posterior Probability:", posterior)
```
